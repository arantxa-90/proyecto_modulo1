```python
#Implementar la palabra que entra en juego:
#Para ello importo la librería random, cada vez que compute esta celda me devuelve una palabra distinta con la que jugar

import random

palabra_secreta = random.choice(palabras_navidad)
palabra_secreta  #solo para verificar en pruebas, luego eliminar
```

```python
#Creación del tablero de juego:

tablero = ['_'] * len(palabra_secreta)
print("\nPalabra a adivinar:")
print(" ".join(tablero))
```

```python
#Pedir letra al jugador y si es válida añadirla a una lista de letras usadas

letras_jugadas = [] #Esta lista se irá actualizando con todas las letras con las que se jueguen, sean correctas o no, para que tengan claro y no repitan
letra = input('¿Con qué letra deseas jugar?: ').upper().strip()

#condiciones de entradas no válidas en el juego
if len(letra) != 1:  #que escriba más de una letra, o una letra y un simbolo...
    print('Error: Debes ingresar solo UNA letra')
elif not letra.isalpha():  #que escriba un número o un símbolo
    print('Error: Solo se permiten letras, no números ni símbolos')
elif letra not in abecedario:
    print('Error: Letra no válida. Usa A-Z sin tildes y sin Ñ.')
elif letra in letras_jugadas:  #que escriba una letra repetida que ya ha usado
    print('Error: Ya ingresaste esa letra, juega con otra diferente.')
else:   #seria la entrada válida, por lo que se añadirá a la lista
    letras_jugadas.append(letra)
    print(f'Las letras que has jugado son: {", ".join(letras_jugadas)}')

```
